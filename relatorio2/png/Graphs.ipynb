```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(font='Franklin Gothic Book', rc={
 'axes.axisbelow': False,
 'axes.edgecolor': 'lightgrey',
 'axes.facecolor': 'None',
 'axes.grid': False,
 'axes.labelcolor': 'dimgrey',
 'axes.spines.right': False,
 'axes.spines.top': False,
 'figure.facecolor': 'white',
 'lines.solid_capstyle': 'round',
 'patch.edgecolor': 'w',
 'patch.force_edgecolor': True,
 'text.color': 'dimgrey',
 'xtick.bottom': False,
 'xtick.color': 'dimgrey',
 'xtick.direction': 'out',
 'xtick.top': False,
 'ytick.color': 'dimgrey',
 'ytick.direction': 'out',
 'ytick.left': False,
 'ytick.right': False})
sns.set_context("notebook", rc={"font.size":16,
                                "axes.titlesize":20,
                                "axes.labelsize":18})
```

```python
# Teste 1
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
x = ["MQTT", "HTTP"]
y = [683, 672]
plt.barh(x, y)
plt.rc('xtick',labelsize=20)
plt.rc('ytick',labelsize=20)
for index, value in enumerate(y):
    plt.text(value+5, index, str(value))
plt.xlabel("\nNúmero total de bytes")
# plt.show()
plt.savefig('teste1.png', bbox_inches='tight')
```

```python
# Teste 2

# Parse HTTP
http = open('../testes/2/http/DATA').readlines()
for i, value in enumerate(http):
    http[i] = float(value)
    
# Parse MQTT    
mqtt = open('../testes/2/mqtt/DATA').readlines()
for i, value in enumerate(mqtt):
    mqtt[i] = float(value)
red_square = dict(markerfacecolor='r', marker='s')
plt.rc('xtick',labelsize=12)
plt.rc('ytick',labelsize=12)
fig1, ax1 = plt.subplots()
# ax1.set_title('Latência em milisegundos')
data = [http, mqtt]
ax1.boxplot(data, flierprops=red_square)
plt.xticks([1, 2], ['HTTP', 'MQTT'])
plt.ylabel("\nMilisegundos", fontsize=12)
# plt.show()
plt.savefig('teste2.png', bbox_inches='tight')
```

```python
# Teste 3

# Parse HTTP
http = open('../testes/3/http/DATA').readlines()
for i, value in enumerate(http):
    http[i] = float(value) / 1000.0

# Parse MQTT    
mqtt = open('../testes/3/mqtt/DATA').readlines()
for i, value in enumerate(mqtt):
    mqtt[i] = float(value) / 1000.0
    
red_square = dict(markerfacecolor='r', marker='s')
plt.rc('xtick',labelsize=12)
plt.rc('ytick',labelsize=12)
fig1, ax1 = plt.subplots()
# ax1.set_title('Latência em milisegundos')
data = [http, mqtt]
ax1.boxplot(data, flierprops=red_square)
plt.xticks([1, 2], ['HTTP', 'MQTT'])
plt.ylabel("\nSegundos", fontsize=12)
# plt.show()
plt.savefig('teste3.png', bbox_inches='tight')
```
