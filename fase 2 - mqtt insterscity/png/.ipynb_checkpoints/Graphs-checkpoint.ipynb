```python
import matplotlib.pyplot as plt
from statistics import mean
import seaborn as sns
sns.set(font='Franklin Gothic Book', rc={
 'axes.axisbelow': False,
 'axes.edgecolor': 'lightgrey',
 'axes.facecolor': 'None',
 'axes.grid': False,
 'axes.labelcolor': 'dimgrey',
 'axes.spines.right': False,
 'axes.spines.top': False,
 'figure.facecolor': 'white',
 'lines.solid_capstyle': 'round',
 'patch.edgecolor': 'w',
 'patch.force_edgecolor': True,
 'text.color': 'dimgrey',
 'xtick.bottom': False,
 'xtick.color': 'dimgrey',
 'xtick.direction': 'out',
 'xtick.top': False,
 'ytick.color': 'dimgrey',
 'ytick.direction': 'out',
 'ytick.left': False,
 'ytick.right': False})
sns.set_context("notebook", rc={"font.size":16,
                                "axes.titlesize":20,
                                "axes.labelsize":18})
```

```python
# Teste 1
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
x = ["MQTT", "HTTP"]
y = [683, 672]
plt.barh(x, y)
plt.rc('xtick',labelsize=20)
plt.rc('ytick',labelsize=20)
for index, value in enumerate(y):
    plt.text(value+5, index, str(value))
plt.xlabel("\nNúmero total de bytes")
# plt.show()
plt.savefig('teste1.png', bbox_inches='tight')
```

```python
# Teste 2

# Parse HTTP
http = open('../testes/2/http/DATA').readlines()
for i, value in enumerate(http):
    http[i] = float(value)
    
# Parse MQTT    
mqtt = open('../testes/2/mqtt/DATA').readlines()
for i, value in enumerate(mqtt):
    mqtt[i] = float(value)
red_square = dict(markerfacecolor='r', marker='s')
plt.rc('xtick',labelsize=12)
plt.rc('ytick',labelsize=12)
fig1, ax1 = plt.subplots()
# ax1.set_title('Latência em milisegundos')
data = [http, mqtt]
ax1.boxplot(data, flierprops=red_square)
plt.xticks([1, 2], ['HTTP', 'MQTT'])
plt.ylabel("\nMilisegundos", fontsize=12)
# plt.show()
plt.savefig('teste2.png', bbox_inches='tight')
```

```python
# Teste 3
x = ['10', '20', '40', '80']
http_y = []

# Parse HTTP
http = open('../testes/3/http/10/DATA').readlines()
for i, value in enumerate(http):
    http[i] = float(value)
http_y.append(round(mean(http)))
http = open('../testes/3/http/20/DATA').readlines()
for i, value in enumerate(http):
    http[i] = float(value)
http_y.append(round(mean(http)))
http = open('../testes/3/http/40/DATA').readlines()
for i, value in enumerate(http):
    http[i] = float(value)
http_y.append(round(mean(http)))
http = open('../testes/3/http/80/DATA').readlines()
for i, value in enumerate(http):
    http[i] = float(value)
http_y.append(round(mean(http)))

# Plot
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
plt.xlabel('Número de threads')
#plt.title('Vazão HTTP', pad=40)
plt.rc('xtick',labelsize=12)
plt.rc('ytick',labelsize=12)
ax.bar(x, http_y, color = 'b')
xlocs, xlabs = plt.xticks()
xlocs=[i+1 for i in range(0,10)]
xlabs=[i/2 for i in range(0,10)]
for i, v in enumerate(http_y):
    plt.text(xlocs[i] - 1.11, v + 5, str(v))


plt.ylabel("\nMensagens por segundo", fontsize=20)
# plt.show()
plt.savefig('teste3_http.png', bbox_inches='tight')
```

```python
# Teste 3
x = ['10', '20', '40', '80']
mqtt_y = []

# Parse MQTT    
mqtt = open('../testes/3/mqtt/10/DATA').readlines()
for i, value in enumerate(mqtt):
    mqtt[i] = float(value)
mqtt_y.append(round(mean(mqtt)))
mqtt = open('../testes/3/mqtt/20/DATA').readlines()
for i, value in enumerate(mqtt):
    mqtt[i] = float(value)
mqtt_y.append(round(mean(mqtt)))
mqtt = open('../testes/3/mqtt/40/DATA').readlines()
for i, value in enumerate(mqtt):
    mqtt[i] = float(value)
mqtt_y.append(round(mean(mqtt)))
mqtt = open('../testes/3/mqtt/80/DATA').readlines()
for i, value in enumerate(mqtt):
    mqtt[i] = float(value)
mqtt_y.append(round(mean(mqtt)))

# Plot
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
plt.xlabel('Número de threads')
#plt.title('Vazão MQTT', pad=40)
plt.rc('xtick',labelsize=12)
plt.rc('ytick',labelsize=12)
ax.bar(x, mqtt_y, color = 'orange')
xlocs, xlabs = plt.xticks()
xlocs=[i+1 for i in range(0,10)]
xlabs=[i/2 for i in range(0,10)]
for i, v in enumerate(mqtt_y):
    plt.text(xlocs[i] - 1.23, v + 600, str(v))
plt.ylabel("\nMensagens por segundo", fontsize=12)
# plt.show()
plt.savefig('teste3_mqtt.png', bbox_inches='tight')
```
